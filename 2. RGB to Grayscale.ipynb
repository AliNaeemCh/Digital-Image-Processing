# RGB to Grayscale

To convert a colored image to grayscale, you typically use a weighted sum of the RGB components to account for human perception of color intensity. The formula used is a linear combination of the red, green, and blue channels. The most common weights, based on luminance perception, are:

$Gray= 0.299 \times R+0.587×G \times 0.114×B$

These weights are chosen because the human eye is more sensitive to green light and less sensitive to blue light.

Input Image Dimension: $y \times x \times 3$

Output Image Dimension: $y \times x \times 1$

Dimension of transformation vector ([0.299; 0.587; 0.114]): $3 \times 1$

Gray Picture = Input Image $\times$ Transformation vector

Gray Picture Dimension = $y \times x \times 1$ $\implies$ $y \times x$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = plt.imread('pictures/road.jpg')
plt.imshow(img)
plt.show()
```

```python
img.shape, img.dtype
```

```python
def to_grayscale(image):
    transformation_vector = np.array([[0.299], [0.587], [0.114]]) # 3 x 1
    ydim, xdim, color_channels = image.shape
    return np.matmul(image, transformation_vector).reshape(ydim, xdim).astype('uint8')
```

```python
img_gray = to_grayscale(img)
plt.imshow(img_gray, cmap='gray', vmin=0, vmax=255)
plt.imsave('pictures/road_gray.jpg', img_gray, cmap='gray', vmin=0, vmax=255)
plt.show()
```
