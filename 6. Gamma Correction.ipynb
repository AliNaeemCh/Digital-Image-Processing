# Gamma Correction
Gamma correction is a non-linear transformation that changes brightness and contrast.


- $\gamma > 1$

When $\gamma > 1$, a lot of pixels are mapped to values closer to zero, giving a darker look.

![Image](pictures/gamma%20correction%20gamma%20gt%20%201.png)

Here $p$ is the normalized pixel value in the range 0 to 1.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = plt.imread('pictures/road_gray.jpg')
img = img[:,:,0]
plt.imshow(img, cmap='gray', vmin=0, vmax=255)
plt.show()
```

```python
gamma = 2
img_norm = img.copy().astype(np.float16)
img_norm = img_norm/255
img_gamma_p = np.power(img_norm, gamma)
img_gamma_p = np.round(img_gamma_p * 255).astype(np.uint8)
```

```python
fig, axs = plt.subplots(1, 2, figsize=(10, 5))

# Original image
axs[0].imshow(img, cmap='gray', vmin=0, vmax=255)
axs[0].set_title('Original Image')
axs[0].axis('off')

# Positive Gamma Corrected image
axs[1].imshow(img_gamma_p, cmap='gray', vmin=0, vmax=255)
axs[1].set_title('Positive Gamma Corrected Image')
axs[1].axis('off')

# plt.savefig('pictures/road_positive_gamma_corrected.jpg')
plt.show()
```

- $\gamma < 1$

When $\gamma < 1$, a lot of pixels are mapped to values closer to 255, giving a brighter look.

![Image](pictures/gamma%20correction%20gamma%20lt%20%201.png)

Here $p$ is the normalized pixel value in the range 0 to 1.

```python
gamma = 0.8
img_norm = img.copy().astype(np.float16)
img_norm = img_norm/255
img_gamma_n = np.power(img_norm, gamma)
img_gamma_n = np.round(img_gamma_n * 255).astype(np.uint8)
```

```python
fig, axs = plt.subplots(1, 2, figsize=(10, 5))

# Original image
axs[0].imshow(img, cmap='gray', vmin=0, vmax=255)
axs[0].set_title('Original Image')
axs[0].axis('off')

# Negative Gamma Corrected image
axs[1].imshow(img_gamma_n, cmap='gray', vmin=0, vmax=255)
axs[1].set_title('Negative Gamma Corrected Image')
axs[1].axis('off')

# plt.savefig('pictures/road_negative_gamma_corrected.jpg')
plt.show()
```
