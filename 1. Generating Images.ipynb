# Generating Images

- The dimension of gray scale image is $y \times x$ (Vertical Axis $\times$ Horizontal Axis)
- The dimension of colored image is $y \times x \times 3$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
xdim = 500
ydim = 300
```

- ### Grayscale constant intensity

```python
img1 = np.full((ydim, xdim), 128, dtype='uint8')
plt.imshow(img1, cmap='gray', vmin=0, vmax=255)
plt.show()
```

- ### Grayscale horizontal stripes

```python
img2 = np.zeros((ydim, xdim), dtype='uint8')
img2[::2, :] = 255
plt.imshow(img2, cmap='gray', vmin=0, vmax=255)
plt.show()
```

- ### Grayscale 2 intensities

```python
img3 = np.full((ydim, xdim), fill_value=64, dtype='uint8')
img3[int(ydim/2):, :] = 192
plt.imshow(img3, cmap='gray', vmin=0, vmax=255)
plt.show()
```

- ### Grayscale ramp

```python
img4 = np.arange(stop=ydim).reshape(ydim, 1)
img4 = np.hstack([img4] * xdim)
img4 = img4/2
img4 = img4.astype('uint8')
plt.imshow(img4, cmap='gray', vmin=0, vmax=255)
plt.show()
```

- ## Grayscale Gaussian Distribution

```python
def gaussian(x, y):
    return 255 * np.exp(-(np.square(x-250)+np.square(y-250))/40000)
```

```python
row_indices, column_indices = np.indices((ydim, xdim))
img5 = gaussian(row_indices, column_indices).astype('uint8')
plt.imshow(img5, cmap='gray', vmin=0, vmax=255)
plt.show()
```

- ## 4 Colors


```python
# Upper Left Quadrant -> Yellow
# Upper Right Quadrant -> Green
# Lower Left Quadrant -> Red
# Lower Right Quadrant -> Black

img6 = np.zeros((ydim, xdim, 3), dtype='uint8')
xhalf = int(xdim/2)
yhalf = int(ydim/2)

# Upper Left Quadrant. R=255, G=255, B=0
img6[:yhalf, :xhalf, 0:2] = 255

# Upper Right Quadrant. R=0, G=255, B=0
img6[:yhalf:, xhalf:, 1] = 255

# Lower Left Quadrant. R=255, G=0, B=0
img6[yhalf:, :xhalf, 0] = 255

# Lower Right Quadrant. R=0, G=0, B=0 (Already zero)

plt.imshow(img6)
plt.show()
```
