# Changing Brightness
Brightness is changed simply by adding to the image values, the offset.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = plt.imread('pictures/road_gray.jpg')
plt.imshow(img)
plt.show()
```

```python
img.shape
```

```python
# Pick any 1 channel, all will be same for a picture which is gray.
img = img[:,:,0]
img.shape
```

- ### Increasing Brightness

By increasing brightness, we may blow out or saturate some pixels as they map to 255 after the brightness increases!

![img](pictures/positive%20offset%20plot.png)

```python
img_brightened = img.copy().astype('int16') + 50
img_brightened = np.clip(img_brightened, 0, 255).astype(np.uint8)

fig, axs = plt.subplots(1, 2, figsize=(10, 5))

# Original image
axs[0].imshow(img, cmap='gray', vmin=0, vmax=255)
axs[0].set_title('Original Image')
axs[0].axis('off')

# Brightened image
axs[1].imshow(img_brightened, cmap='gray', vmin=0, vmax=255)
axs[1].set_title('Brightened Image')
axs[1].axis('off')

# plt.savefig('pictures/road_brightened.jpg')
plt.show()
```

- ### Decreasing Brightness

By increasing brightness, we may blowout or saturate some pixels as they map to 0 after the brightness decreases!

![img](pictures/negative%20offset%20plot.png)

```python
img_darkened = img.copy().astype('int16') - 50
img_darkened = np.clip(img_darkened, 0, 255).astype(np.uint8)

fig, axs = plt.subplots(1, 2, figsize=(10, 5))

# Original image
axs[0].imshow(img, cmap='gray', vmin=0, vmax=255)
axs[0].set_title('Original Image')
axs[0].axis('off')

# Darkened image
axs[1].imshow(img_darkened, cmap='gray', vmin=0, vmax=255)
axs[1].set_title('Darkened Image')
axs[1].axis('off')

# plt.savefig('pictures/road_darkened.jpg')
plt.show()
```
